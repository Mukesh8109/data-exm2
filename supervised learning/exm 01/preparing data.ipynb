# Final Conclusion

The objective of this analysis was to explore the house price dataset and prepare it for machine learning modeling.

Initially, the dataset was loaded and inspected to understand the available variables. The dataset contains features such as area_sqft, bedrooms, bathrooms, location_score, age_years, distance_city_km, lot_size_sqft, has_garage, has_pool, and renovation_years_ago, with house_price_inr as the target variable.

## Feature and Target Separation

The dataset was divided into:

Independent variables (X): All property-related features excluding house_price_inr and house_id.

Dependent variable (y): house_price_inr, which represents the house price to be predicted.

## Bivariate Analysis

Scatter plots were created to analyze the relationship between each feature and the target variable. The analysis indicates that variables such as area_sqft, bedrooms, bathrooms, location_score, and lot_size_sqft show a positive relationship with house prices, meaning that houses with larger sizes, more rooms, and better locations tend to have higher prices.

On the other hand, features like distance_city_km and age_years show a negative relationship with house prices, suggesting that houses farther from the city center or older properties generally have lower prices.

A box plot was also used to examine how the number of bedrooms influences house prices, showing that houses with more bedrooms generally fall into higher price ranges.

## Correlation Analysis

A correlation heatmap was generated to understand the relationships among all numerical variables. The heatmap helped identify which features are strongly associated with house prices and which variables may influence each other.

## Data Preparation for Machine Learning

Finally, the dataset was split into training and testing sets using an 80–20 ratio with train_test_split. This step prepares the data for building and evaluating machine learning models.

# Overall Insight

The exploratory analysis indicates that property size, number of rooms, location quality, and lot size are major factors influencing house prices, while distance from the city and property age tend to reduce price value.

With the dataset now properly explored and split into training and testing sets, it is ready for the next stage of machine learning model development and price prediction.
