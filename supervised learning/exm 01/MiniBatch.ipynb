```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('D:\AI&ML\Supervised Learning\Supervised Learning Task\Supervised-Learning\Exam 1\RealEstate Dataset.csv')
```

```python
df.head()
```

```python
X = df.drop("house_price_inr", axis=1).values
y = df["house_price_inr"].values.reshape(-1,1)
```

```python
# Feature normalization
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Add bias column
X = np.c_[np.ones(X.shape[0]), X]
```

```python
n_samples, n_features = X.shape
theta = np.zeros((n_features,1))
```

```python
learning_rate = 0.01
epochs = 100
batch_size = 32
```

```python
def mini_batch_gradient_descent(X, y, theta, lr, epochs, batch_size):
    
    n_samples = X.shape[0]

    for epoch in range(epochs):

        # Shuffle data
        indices = np.random.permutation(n_samples)
        X_shuffled = X[indices]
        y_shuffled = y[indices]

        for i in range(0, n_samples, batch_size):

            X_batch = X_shuffled[i:i+batch_size]
            y_batch = y_shuffled[i:i+batch_size]

            m = len(X_batch)

            # Predictions
            y_pred = X_batch @ theta

            # Gradient
            gradient = (1/m) * X_batch.T @ (y_pred - y_batch)

            # Update weights
            theta = theta - lr * gradient

    return theta

theta = mini_batch_gradient_descent(X, y, theta, learning_rate, epochs, batch_size)

print("Trained Weights:")
print(theta)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X[:,1:], y)

print("Sklearn Weights:", model.coef_)
print("Your Weights:", theta[1:])
```

Conclusion

The Mini-Batch Gradient Descent method was implemented to train a linear regression model for predicting house prices using multiple housing features. Instead of updating the model parameters using the entire dataset or a single sample, the algorithm updates the weights using small batches of data. This approach improves computational efficiency and provides faster convergence compared to traditional batch gradient descent.

After training the model and evaluating it using regression metrics such as Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R² score, the results indicate that the model is able to learn the relationship between the housing features and the target price. The evaluation metrics and prediction results confirm that the Mini-Batch Gradient Descent algorithm successfully optimized the model parameters.

Overall, Mini-Batch Gradient Descent provides a balanced approach between computational efficiency and model stability, making it a suitable optimization technique for training machine learning models on moderate to large datasets.
