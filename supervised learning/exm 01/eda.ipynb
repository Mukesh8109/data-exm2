## Outlier

```python
for col in num_cols:
    sns.boxplot(df[col])
    plt.title(f"Outlier Handling {col}")
    plt.show()
```

Outlier Present columns it's to much.

#### Outlier Handling

```python
new_df= df.copy()

for col in num_cols:
    lower = df[col].quantile(0.05)
    upper = df[col].quantile(0.95)
    new_df[col] = np.clip(new_df[col], lower, upper)
new_df
```

Outlier Value replace by the boader values.

```python
for col in num_cols:
    sns.boxplot(new_df[col])
    plt.title(f"Outlier Handling {col}")
    plt.show()
```

```python
for col in num_cols:
    sns.histplot(new_df[col],kde=True)
    plt.show()
```

# Final Conclusion

### Univariate Analysis

Univariate analysis was performed using histograms and KDE plots to understand the distribution of each numerical feature. This helped identify patterns such as normal distribution, skewness, and potential outliers. Skewness was further evaluated using .skew() to detect features that deviate from normal distribution.

### Bivariate Analysis

Bivariate analysis was conducted using a correlation heatmap, which helped identify relationships between numerical variables. This analysis helped determine which variables may influence others and which features may be important for predictive modeling.

### Outlier Treatment

To reduce the effect of extreme values, Winsorization was applied using the 5th and 95th percentiles. Instead of removing outliers, extreme values were capped using numpy.clip(), preserving the dataset size while minimizing the impact of abnormal values.

### Normality Checking

Q-Q plots were generated for numerical variables to compare their distributions with a theoretical normal distribution. This step helped identify whether transformations may be needed for modeling techniques that assume normality.

### Automated Data Profiling
A data profiling report was also generated using the ydata_profiling library, which provided a comprehensive overview of the dataset, including distributions, correlations, missing values, and warnings.

### Overall Findings

The exploratory analysis provided valuable insights into the dataset, including:

The distribution of numerical variables

Relationships between features

Data quality and structure
